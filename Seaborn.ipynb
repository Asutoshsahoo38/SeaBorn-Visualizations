```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")## Here tips is the dependent feature and rest all are independent i.e 
##tips depend on all other independent features
df.head()
```

```python
#Correlation with heatmap
df.corr()# correlation between two features
#And correlation can only be applied with only integer and float values
```

```python
sns.heatmap(df.corr())##the color shading will show the relation between the features 
```

Joint plot

```python
sns.jointplot(x = 'tip' , y = 'total_bill',data = df )##it will just sjow the relationship between the two 
#features i.e tip and total_bill
```

```python
sns.jointplot(x = 'tip' , y = 'total_bill',data = df , kind = 'hex')#it will show the relationship of the two 
# features in hexagonal way 
```

```python
sns.jointplot(x = 'tip' , y = 'total_bill',data = df,kind = 'reg' )#it will show the best regration line between 
#two features with plotting a line
```

pair plot

```python
sns.pairplot(df)##it will just perform the permutation and combination of the data.All data will be shown in scatter
#and histogram order
```

```python
sns.pairplot(df,hue = 'sex')##All the plot are drawn on the basic of sex
```

Dist plot

```python
sns.distplot(df['tip'])##it will show a histogram on the basic of tip and its density
```

```python
sns.distplot(df['tip'],kde = False,bins = 10)##If kde will false it will show the tips with respect to count
##other wise it will show in density
##bins is the difference between the count of the numbers
```

categorical plot
->Box plot
->violin plot
->count plot
->Bar plot

```python
#count plot
sns.countplot(y = 'sex',data = df)#it will just count the no. of occurance 
```

```python
sns.countplot(x = 'smoker',data = df)#in count we have to take any single features x or y
```

```python
#Bar plot
#here we have to take both x and y
sns.barplot(x='total_bill',y='sex',data = df)
```

```python
sns.barplot('tip', 'total_bill',data = df)
```

```python
#box plot
sns.boxplot(x = 'smoker',y='total_bill',data =df)
```

```python
sns.boxplot('day','total_bill',data = df,palette = 'rainbow')#palette will show the plots in rainbow colours
```

```python
sns.boxplot('day','total_bill',hue = 'smoker',data = df)
```

```python
#violin plot
#violin plot helps us to see both distribution of the data i.e kernel density and the box plot
sns.violinplot('total_bill','day',data=df,palette='rainbow')
#internal is the box plot
#it will show the distribution of total_bill and day with respect to data
```
